## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

**Why ML Beats Fixed Rules:**
A fixed rule-based approach (e.g., if content_ctr > 0.10 then rank_high) fails because real-world user behavior involves non-linear interactions across multiple noisy variables (user historical preference, session context, time decay, content freshness). ML models continuously adapt to subtle non-linear signals and contextual shifts without requiring endless manual, fragile if/else conditions.
